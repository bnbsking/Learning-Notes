```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
# Linear Regression
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
# 1 data
train_x = np.array([[-1], [-2], [-3], [1], [2], [3]])
train_y = np.dot(train_x, np.array([2])) + 3  # y=2x+3
valid_x = np.array([[2.5]])
valid_y = np.dot(valid_x, np.array([2])) + 3  # y=2x+3
print("train_x.shape=", train_x.shape, "train_y.shape=", train_y.shape)

# 2 model
reg = LinearRegression().fit(train_x, train_y)

# 3 predict
ypred = reg.predict(valid_x)
print("pred_y=", ypred)
print("r2_score_train=", reg.score(train_x, train_y))  # R^2, 0~1
print("r2_score_valid=", reg.score(valid_x, valid_y))  # R^2, 0~1

# 4 plot
plt.scatter(train_x, train_y)
plt.scatter(valid_x, valid_y, color='green')
plt.scatter(valid_x, ypred, color='red', marker='x')
plt.grid()
plt.show()
```

```python
# Support Vector Classification
# https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
# 1 data
train_x = np.array([[-1,-1], [-2,-1], [-3,-2], [1,1], [2,1], [3,2], [-2,1], [-3,2], [-1,2]])
train_y = [0, 0, 0, 1, 1, 1, 2, 2, 2]
valid_x = np.array([[2, 2]])
valid_y = [1]

# 2 model
svc = SVC().fit(train_x, train_y)  # kernel in {‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’}, default=’rbf’

# 3 predict
ypred = svc.predict(valid_x)
print("ypred=", ypred)
print("accuracy_train=", svc.score(train_x, train_y))  # 0~1
print("accuracy_valid=", svc.score(valid_x, valid_y))  # 0~1

# 4 plot
color = ["blue", "green", "red"]
plt.scatter(train_x[:,0], train_x[:,1], c=[color[y] for y in train_y])
plt.scatter(valid_x[:,0], valid_x[:,1], c=[color[y] for y in valid_y], marker='x')
plt.scatter(valid_x[:,0], valid_x[:,1], c=[color[y] for y in ypred], marker='*')
plt.grid()
plt.show()
```

```python
# Clustering
# https://scikit-learn.org/1.5/modules/generated/sklearn.cluster.KMeans.html
# 1 data
train_x = np.array([[1,2], [2,3], [4,2], [3,3], [-2,-2], [-1,-3], [-3,-4], [-2,-4]])

# 2 model
kmeans = KMeans(n_clusters=2).fit(train_x)

# 3 predict
centers = kmeans.cluster_centers_
print(f"centers=", centers)
pred = kmeans.predict(train_x)
print("pred=", pred)

# 4 plot
plt.scatter(train_x[:,0], train_x[:,1], c=pred)
plt.show()
```

```python
# Dimensionality Reduction
# https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html
# 1 data
train_x = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])

# 2 model
reduction = TSNE(n_components=1, perplexity=3).fit_transform(train_x)
reduction = (reduction-reduction.mean())/reduction.std()

# 3 predict
print("reduction=", reduction)

# 4 plot
plt.scatter(train_x[:,0], train_x[:,1])
plt.scatter(reduction, [0] * len(train_x))
plt.grid()
plt.show()
```
